## 8. Ingienier√≠a de caracter√≠sticas

### Data Cleaning

```python
import numpy as np
from sklearn.impute import SimpleImputer

# Eliminar duplicados exactos
df = df.drop_duplicates().reset_index(drop=True)

# Tratar nulos:
# Separar columnas num√©ricas y categ√≥ricas
num_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()
cat_cols = df.select_dtypes(include=['object','category']).columns.tolist()
# Excluir ID si est√° en num_cols
if 'student_id' in num_cols: num_cols.remove('student_id')
if 'course_id' in num_cols: num_cols.remove('course_id')

# Imputaci√≥n num√©rica (mediana) y categ√≥rica (modo)
num_imputer = SimpleImputer(strategy='median')
cat_imputer = SimpleImputer(strategy='most_frequent')

df[num_cols] = num_imputer.fit_transform(df[num_cols])
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

# Corregir inconsistencias por fuzzy matching (ejemplo simplificado)
# Ej: corregir 'Argentin' -> 'Argentina', 'bachlor' -> 'Bachelor'
df['country'] = df['country'].replace({'Argentin': 'Argentina'})
df['education_level'] = df['education_level'].replace({'bachlor': 'Bachelor'})

# Outliers simples: winsorizaci√≥n ejemplo en payment_amount_usd
q_low = df['payment_amount_usd'].quantile(0.01)
q_high = df['payment_amount_usd'].quantile(0.99)
df['payment_amount_usd'] = np.clip(df['payment_amount_usd'], q_low, q_high)

# Valores imposibles de time_spent_hours (menores a 0)
df.loc[df['time_spent_hours'] < 0, 'time_spent_hours'] = df['time_spent_hours'].median()

print("Cleaning done. New shape:", df.shape)
```

### Feature Engeneering

```python
# Feature engineering (ejemplos)
# 4.1 Engagement score simple (normalizar entradas)
df['engagement_score'] = (df['time_spent_hours']/df['time_spent_hours'].max() +
                          df['videos_watched']/df['videos_watched'].max() +
                          df['assignments_submitted']/ (df['assignments_submitted'].max()+1)) / 3

# 4.2 Academic success composite
df['academic_success'] = (df['project_grade'].fillna(0) + df['final_exam_score'].fillna(0)) / 2

# 4.3 Bucketizar edad
df['age_group'] = pd.cut(df['age'], bins=[15,24,34,44,60,120], labels=['16-24','25-34','35-44','45-60','60+'])

# 4.4 Tiempo relativo de completitud
df['fast_completion'] = (df['days_to_complete'] < df['days_to_complete'].median()).astype(int)

# 4.5 One-Hot de algunas categ√≥ricas si se necesitan luego
# (lo haremos en Pipeline para modelos)
df.head()
```

### 9. Metodolog√≠a y Modelos Utilizados


```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer # Although imported before, good to have it here for clarity

# Preparar datos para modelado (churn prediction)

# Selecciono features y target
TARGET = 'churn'   # problema de clasificaci√≥n binaria
features = ['age','income_level','subscription_type','course_category','engagement_score',
            'academic_success','discount_used','marketing_channel','payment_amount_usd','age_group']

X = df[features].copy()
y = df[TARGET].astype(int).copy()

# Train-test split estratificado
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Columnas por tipo para el pipeline
num_features = ['age','engagement_score','academic_success','payment_amount_usd']
cat_features = ['income_level','subscription_type','course_category','marketing_channel','age_group','discount_used']

# Pipelines
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', num_pipeline, num_features),
    ('cat', cat_pipeline, cat_features)
])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, roc_auc_score

# Modelo Supervisado (Clasificaci√≥n) ‚Äî Baseline + RF + evaluaci√≥n

# Baseline con LogisticRegression dentro de un pipeline
pipe_lr = Pipeline([
    ('pre', preprocessor),
    ('clf', LogisticRegression(max_iter=1000))
])
pipe_lr.fit(X_train, y_train)
y_pred = pipe_lr.predict(X_test)
y_proba = pipe_lr.predict_proba(X_test)[:,1]
print("Logistic Regression - Classification Report")
print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))

# RandomForest con GridSearch (ejemplo peque√±o)
pipe_rf = Pipeline([
    ('pre', preprocessor),
    ('clf', RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1))
])

param_grid = {
    'clf__n_estimators': [100, 200],
    'clf__max_depth': [5, 10, None],
    'clf__min_samples_split': [2,5]
}

cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=42)
grid_rf = GridSearchCV(pipe_rf, param_grid, cv=cv, scoring='roc_auc', n_jobs=-1, verbose=1)
grid_rf.fit(X_train, y_train)
print("Best params:", grid_rf.best_params_)
best_rf = grid_rf.best_estimator_

y_pred_rf = best_rf.predict(X_test)
y_proba_rf = best_rf.predict_proba(X_test)[:,1]
print("RF Classification Report")
print(classification_report(y_test, y_pred_rf))
print("RF ROC AUC:", roc_auc_score(y_test, y_proba_rf))
```

```python
# Manejo de desequilibrio (SMOTE) - opcional dentro de pipeline
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
imb_pipeline = ImbPipeline([
    ('pre', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('clf', RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1))
])

imb_pipeline.fit(X_train, y_train)
y_pred_sm = imb_pipeline.predict(X_test)
print(classification_report(y_test, y_pred_sm))
print("ROC AUC (SMOTE):", roc_auc_score(y_test, imb_pipeline.predict_proba(X_test)[:,1]))
```

```python
# Modelos de regresi√≥n (ejemplo: predecir satisfaction_rating)
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np # Import numpy for np.sqrt

TARGET_REG = 'satisfaction_rating'
features_reg = ['engagement_score','academic_success','payment_amount_usd','subscription_type','course_category','age']

Xr = df[features_reg].copy()
yr = df[TARGET_REG].astype(float).copy()

Xr_train, Xr_test, yr_train, yr_test = train_test_split(Xr, yr, test_size=0.2, random_state=42)

num_features_r = ['engagement_score','academic_success','payment_amount_usd','age']
cat_features_r = ['subscription_type','course_category']

preprocessor_r = ColumnTransformer([
    ('num', Pipeline([('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())]), num_features_r),
    ('cat', Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))]), cat_features_r)
])

pipe_reg = Pipeline([
    ('pre', preprocessor_r),
    ('rf', RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1))
])
pipe_reg.fit(Xr_train, yr_train)
yr_pred = pipe_reg.predict(Xr_test)
print("RMSE:", np.sqrt(mean_squared_error(yr_test, yr_pred)))
```

```python
# No supervisado ‚Äî Clustering (segmentaci√≥n de estudiantes)
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# PCA + KMeans para segmentaci√≥n por engagement y behavior
cluster_features = ['engagement_score','academic_success','payment_amount_usd','videos_watched','assignments_submitted']
Xc = df[cluster_features].fillna(0)

scaler = StandardScaler()
Xc_scaled = scaler.fit_transform(Xc)

# PCA para visualizaci√≥n
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(Xc_scaled)

# Buscar k con elbow/simple silhouette
from sklearn.metrics import silhouette_score
scores = {}
for k in range(2,7):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(Xc_scaled)
    scores[k] = silhouette_score(Xc_scaled, labels)
scores

# Elegimos k √≥ptimo (ej: 3)
k = max(scores, key=scores.get)
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(Xc_scaled)

# Visualizar
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=df['cluster'], palette='tab10')
plt.title(f'Clusters k={k}')
plt.show()

# Resumen por cluster
display(df.groupby('cluster')[cluster_features + ['churn','upgraded_membership']].mean())
```

```python
# Aprendizaje por refuerzo (propuesta simple: contextual bandit - epsilon-greedy)

# Eesqueleto sencillo para simular una pol√≠tica de descuentos que aprende con un enfoque epsilon-greedy (√∫til para prototipado antes de pasar a RL m√°s serio).

# Contextual bandit simple:
# Context: engagement_score + academic_success + subscription_type encoded
# Actions: [0% discount, 10% discount, 25% discount]
actions = [0, 10, 25]
n_actions = len(actions)

# Preparar contexto X_bandit (np.array)
context_cols = ['engagement_score','academic_success','payment_amount_usd']
X_bandit = df[context_cols].fillna(0).values
n_rounds = X_bandit.shape[0]

# Inicializar valores Q por acci√≥n (promedio de reward)
Q = np.zeros(n_actions)
N = np.zeros(n_actions)  # conteo de veces que se eligi√≥ acci√≥n

def simulate_reward(action, idx):
    """Funci√≥n que simula reward: si usuario compra/upgrade luego del descuento.
       Aqu√≠ definimos heur√≠stica: mayor engagement + descuento adecuado aumenta prob de compra."""
    base_prob = 0.02 + 0.5 * X_bandit[idx,0]  # engagement principal
    # penalizamos descuentos muy altos si usuario ya pago mucho
    price_factor = 1 - min(X_bandit[idx,2]/300.0, 0.8)
    discount_effect = 0.01 * actions[action]  # cada punto % aporta
    prob = np.clip(base_prob * price_factor + discount_effect*0.02, 0, 1)
    return np.random.binomial(1, prob)  # 0/1 reward (compra o no)

# Epsilon-greedy run
epsilon = 0.1
for t in range(n_rounds):
    if np.random.rand() < epsilon:
        a = np.random.randint(0, n_actions)
    else:
        a = np.argmax(Q)
    r = simulate_reward(a, t)
    N[a] += 1
    Q[a] += (r - Q[a]) / N[a]

print("Estimated Q values:", Q)
print("Action counts:", N)
# Resultado: ver cu√°l descuento genera mejor reward en este entorno simulado
```
